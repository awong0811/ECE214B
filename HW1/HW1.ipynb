```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = r'HW1_audio_files\wekeptourmoney.wav'  # Replace with your file's path
y, sr = librosa.load(file_path, sr=None)  # sr=None preserves the original sampling rate

# `y` contains the audio time-series data (samples)
# `sr` is the sampling rate of the audio
print(f"Audio data shape: {y.shape}")
print(f"Sampling rate: {sr}")

start_time = 1.72
end_time = 3.07
start_sample = int(start_time * sr)
end_sample = int(end_time * sr)
y = y[start_sample:end_sample]
print(f"Cropped audio data shape: {y.shape}")
```

```python
# Compute the Short-Time Fourier Transform (STFT)
D = librosa.stft(y)

# Convert the complex-valued STFT to a magnitude spectrogram
S, phase = librosa.magphase(D)

# Convert to a decibel scale for better visualization
S_dB = librosa.amplitude_to_db(S, ref=np.max)

# Plot the spectrogram
plt.figure(figsize=(10, 6))
librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='log', cmap='magma')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram (Normal) (dB)')
plt.xlabel('Time')
plt.ylabel('Frequency (log scale)')
plt.show()
```

```python
# Extract pitch (fundamental frequency) using pYIN
fmin = librosa.note_to_hz('C2')  # Minimum frequency, e.g., C2 (~65.4 Hz)
fmax = librosa.note_to_hz('C7')  # Maximum frequency, e.g., C7 (~2093 Hz)
f0, voiced_flag, voiced_prob = librosa.pyin(y=y, fmin=fmin, fmax=fmax, sr=sr)

# Generate the time axis for the pitch values
times = np.linspace(start=0, stop=end_time-start_time, num=len(f0), endpoint=True)

# Plot the pitch over time
plt.figure(figsize=(10, 6))
plt.plot(times, f0, label='Pitch', color='b')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('Pitch vs. Time (Normal)')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```

```python
rms_energy = librosa.feature.rms(y=y)

# Generate the time axis for the RMS energy
times = np.linspace(start=0, stop=end_time-start_time, num=len(f0), endpoint=True)

# Plot the energy over time
plt.figure(figsize=(10, 6))
plt.plot(times, rms_energy[0], label='Energy', color='g')
plt.xlabel('Time (s)')
plt.ylabel('Energy (RMS)')
plt.title('Energy vs. Time (Normal)')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```
